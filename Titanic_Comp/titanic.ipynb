```python
#handling missing value
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
df.drop(columns=['Cabin'], inplace=True)
```

```python
#feature extraction
df['family_size']=df['SibSp']+df['Parch']+1
```

```python
df.head(10)
```

```python
#converting categorical variable into numerical 
df=pd.get_dummies(df,columns=['Sex','Embarked'],drop_first=True)    
```

```python
X=df.drop(columns=['PassengerId','Survived','Name','Ticket'])
y=df['Survived']                
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=1000)
model.fit(X,y)
```

```python
test=pd.read_csv('test.csv')
test.head()
```

```python
test['family_size']=test['SibSp']+test['Parch']+1
test['Age'].fillna(test['Age'].mean(),inplace=True)
test['Fare'].fillna(test['Fare'].median(),inplace=True)
test['Embarked'].fillna(test['Embarked'].mode()[0],inplace=True)
```

```python
text=pd.get_dummies(test,columns=['Sex','Embarked'],drop_first=True)
```

```python
test=test.drop(columns=['PassengerId','Name','Ticket'])
```

```python
test =test.reindex(columns=X.columns, fill_value=0)
```

```python
X_test = test
y=test['Survived']
```

```python
predict=model.predict(test)
predict
```

```python
X_1 = df[['Fare']] 
```

```python
X_range = np.linspace(X_1.min(), X_1.max(), 300).reshape(-1, 1)
y_prob = model.predict_proba(X_range)[:, 1]

plt.scatter(X_1, y, alpha=0.3)
plt.plot(X_range, y_prob)
plt.xlabel("Fare")
plt.ylabel("Probability of Survival")
plt.title("Logistic Regression Curve")
plt.show()

plt.show()
```
